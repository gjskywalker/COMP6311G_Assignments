```python
# --- Import Visualization Libraries ---
import matplotlib.pyplot as plt
import numpy as np
```

```python
# --- Define Input Sizes and Calculate Complexity ---
n = np.linspace(1, 100, 100)

linear_complexity = n
tree_complexity = np.log2(n)
```

```python
# --- Generate Time Complexity Plot ---
plt.figure(figsize=(10, 3))

plt.plot(n, linear_complexity, label='Linear Search $O(N)$', color='red', linestyle='--')
plt.plot(n, tree_complexity, label='Tree Index Search $O(\log N)$', color='blue', linewidth=2)

plt.title('Time Complexity Comparison: Linear vs. Tree Search', fontsize=18)
plt.xlabel('Input Size ($N$)', fontsize=16)
plt.ylabel('Theoretical Operations', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=16)
plt.grid(True, alpha=0.3)

plt.show()
plt.savefig('time_complexity_comparison.pdf', dpi=300)
```
